# Import the usual Packages

```python
import numpy as np
import matplotlib.pyplot as plt 
```

# Read the Comsol Text File

```python
filename = 'data.txt'
data = np.loadtxt(filename, comments='%')
```

# Print the number of columns and the number of rows

```python
nrows, ncols = data.shape
print('rows:',nrows)
print('cols:',ncols)
```

# Get the x and y coordinates

Assume that this is regularly spaced data. The two resulting variables

x = ....
y = ....

should be 1D arrays, similar to what you would get from a numpy.linspace command

```python
x = np.unique(data[:,0])
y = np.unique(data[:,1])
```

# With these two 1D arrays , compute the space between points of evaluation

```python
all_x_spacings = np.diff(x)
all_y_spacings = np.diff(y)

x_spacing = all_x_spacings[0]
y_spacing = all_y_spacings[0]
```

Extra task : ensure that the distance between the grid points is really constant 

```python
print(np.allclose(all_x_spacings, x_spacing))
print(np.allclose(all_y_spacings, y_spacing))
```

# Create 2D arrays for the coordinates using meshgrid, initialize the two displacement fields: 

xx , yy = np.meshgrid(x,y)
u = np.zeros_like(xx)
v = np.zeros_like(yy)

## and store the displacements data from data in the 2D arrays u and v

Hint 1 : one way of doing this is going row by row through the data array and extract for each of these rows the coordinates and the displacements

Hint 2 : Think about , if the data points are the values of cell centers or the values of cell corners.

Hint - question : How can you get the indices that you need for the 2D x/y displacement array

```python
xx , yy = np.meshgrid(x, y)
u = np.zeros_like(xx)
v = np.zeros_like(yy)

for row in data:
    x_value = row[0] - x_spacing/2
#     print(x_value)
    y_value = row[1] - y_spacing/2
#     print(y_value)
    u_value = row[2]
    v_value = row[3]
    
    i = int(x_value/x_spacing)
#     print(x_spacing)
#     print(i)
    j = int(y_value/y_spacing)
#     print(y_spacing)
#     print(j)
    u[j,i] = u_value
    v[j,i] = v_value
```

# 7. Plot the displacement fields

use contourf to plot the x and y displacement data and add a colorbar

Think about , if the data points are the values of cell corners or the values of cell corners and use xlim/ylim to set the plot ranges

Add x- and y-labels , as well as a title for each plot 

What is a suitable color map for the displacement data? 

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, figsize=(15,7))
levels = np.linspace(-5e-2, 5e-2, 17)

cf = ax0.contourf(xx, yy, u, cmap='RdBu', levels=levels)
ax0.set(xlim=(0,1), ylim=(0, 0.3), xlabel = 'x [m]', ylabel = 'y [m]', title = 'displacement u', aspect = 'equal')
plt.colorbar(cf, ax = ax0, aspect = 12)

cf = ax1.contourf(xx, yy, v, cmap='RdBu', levels=levels)
ax1.set(xlim=(0,1), ylim=(0, 0.3), xlabel = 'x [m]', ylabel = 'y [m]', title = 'displacement v', aspect = 'equal')
plt.colorbar(cf, ax = ax1, aspect = 12)

plt.tight_layout()
```
